```python
# 5. CLEANING INDIVIDUAL COLUMNS- ii)  column instant_bookable  - Let's say we are going to train a machine learning model & what we need to do 
# in machine learning model is we need to have numeric values for training the model so there's only True and False values. 
```

```python
data["instant_bookable"]
```

```python
# we use apply() . it's going to go rowwise so i will say lamda X one  let's say it's going to be 1 if it is True and 0 for False 
data["instant_bookable"].apply( lambda  x:1 if x== True else 0 )
```

```python
# Explanation of above :
#The lambda keyword is used to create anonymous functions (functions without a name) in Python.
#The syntax for a lambda function is: lambda arguments: expression.
#In this case, x is the argument, and 1 if x == True else 0 is the expression that gets evaluated.
#The apply() method in pandas is used to apply a function (in this case, the lambda function) along the axis of a DataFrame or Series.
#In essence, this code iterates through the "instant_bookable" column of your data DataFrame. For each value (x) in that column,
#it checks if x is equal to True. 
#If it is, it returns 1; otherwise, it returns 0. The resulting series of 0s and 1s replaces the original "instant_bookable" column.
```

```python
data
```

```python
#here above data is still showing True or False value in column instant_bookable . so we need to modify the above code.
```

```python
data["instant_bookable"]= data["instant_bookable"].apply( lambda  x:1 if x== True else 0 )
```
